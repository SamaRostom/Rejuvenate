```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
import argparse
from imutils import paths
import cv2
```

```python
#out is the folder that will contain the augmented images
datapath="/content/drive/My Drive/Non-cleft"
out="/content/drive/My Drive/augimages/target"
```

```python
print("[INFO] loading images...")
imagePaths = list(paths.list_images(datapath))
print(imagePaths)
data=[]
```

```python
#resize the images and convert color to RGB and put it in datanp array to use it
for imagePath in imagePaths:
    print(imagePath)
    image = cv2.imread(imagePath)
	  #cv2.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    print(image)
    print("////////////////////////")
    image = img_to_array(image)
    print(image)
    image = np.expand_dims(image, axis=0)
    data.append(image)
datanp=np.array(data)
```

```python
# ImageDataGenerator class ensures that the model receives new variations of the images
print(len(datanp))
aug = ImageDataGenerator(
	rotation_range=30,
	zoom_range=0.15,
	width_shift_range=0.2,
	height_shift_range=0.2,
	shear_range=0.15,
	horizontal_flip=True,
	fill_mode="nearest")
total = 0
imageGen=[]
```

```python
# apply image augmentation on the images
print("[INFO] generating images...")

for i in range(len(datanp)):
  print("Generating is running")
  imageGen = aug.flow(datanp[i], batch_size=1,save_to_dir=out, save_prefix="image", save_format="jpg")
  for image in imageGen:
    total+=1
    if (total % 6 == 0):
      break
```
